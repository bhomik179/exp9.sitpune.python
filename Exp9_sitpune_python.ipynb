```python

import numpy as np

# 1‑D array
a = np.array([1, 10, 20, 30, 40, 50, 500])
print()

# 2‑D array
b = np.array([[1, 2, 3],
               [4, 5, 6]])

print("a =", a)
print("b =")
print(b)
print()

#Shape (dimensions)
print("a.shape =", a.shape)
print("b.shape =", b.shape)
print()

#Number of dimensions
print("a.ndim =", a.ndim)
print("b.ndim =", b.ndim)
print()

#Data type of the arrays
print("a.dtype =", a.dtype)
print("b.dtype =", b.dtype)
print()

#Size of each element in bytes
print("a.itemsize =", a.itemsize)
print("b.itemsize =", b.itemsize)
print()


# Zero matrix (3×3)
e = np.zeros((3, 3))
print("zeros matrix:\n", e)
print()

u = np.eye(3)
print("identity matrix:\n", u)
print()

# Creating a range of numbers (start, stop, step)
c = np.arange(1, 10, 2)   # 1, 3, 5, 7, 9
print("arange(1,10,2) =", c)
print()
# linespace

h=np.linspace(0,10,5)
print("linspace(0,10,5) =", h)
print()


# adding adn multipying the into the same matrix
print("Adding with the number 5 in matrix\n",a+5)
print("Multiplying with the number 2 in matrix\n",b*2)
print()

mean_val = np.mean(a)

#median
median_val = np.median(a)

print("The value of the mean   is", mean_val)
print("The value of the median is\n", median_val)

print()


#maximum
max_val = np.max(a)

print("Maximum value:", max_val)
print()

# minimum
min_val = np.min(a)
print("Minimum value:", min_val)
print()

# sum
sum_val = np.sum(a)
print("Sum of all values:", sum_val)
print()












```
